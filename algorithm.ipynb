```python
from sklearn.cluster import KMeans

def clus_en_matrix(X, L, K):
    n = len(X)
    c_mem = np.zeros((n,L), dtype=int)
    A_m = [None]*L
    K_l = np.zeros(L, dtype=int)
    D = csr_matrix((n,n), dtype='int')
#     opts = ...
    for l in range(L):
        C = KMeans(n_clusters=K+l, random_state=0).fit(X)
        c_mem[:,l] = C.labels_
        
    for l in range(L):
        A_m[l] = csr_matrix((np.ones(n, dtype='int'),
                    (np.arange(n),c_mem[:,l])), shape=(n,K+l))
        
    A = A_m[1]
    for l in range(1,L):
        A = pd.concat([A, A_m[l]], axis=1)
    
    A = np.sqrt(1/L)*A
    return A
```
